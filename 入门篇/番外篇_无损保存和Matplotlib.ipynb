```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('test.jpg')
img2 = img[:, :, :: -1] #  BGR -> RGB
# 
# img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# 
plt.subplot(1, 2, 1)
plt.imshow(img)

# 
plt.subplot(1, 2, 2)
plt.xticks([]), plt.yticks([]) # x, y
plt.imshow(img2)

plt.show()
```
