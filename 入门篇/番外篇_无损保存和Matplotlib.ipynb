## 显示彩色图

OpenCV中的图像使以BGR的通道顺序存储的，但Matplotlib是以RGB模式显示的，所以需要进行转换

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('test.jpg')
img2 = img[:, :, :: -1] # 颜色翻转 BGR -> RGB
# 或使用
# img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# 显示不正确的图
plt.subplot(1, 2, 1)
plt.imshow(img)

# 显示正确的图
plt.subplot(1, 2, 2)
plt.xticks([]), plt.yticks([]) # 隐藏x, y轴
plt.imshow(img2)

plt.show()
```

## Matplotlib加载和保存图片

加载图片imread

保存图片savefig

```python
import matplotlib.image as pli

img = pli.imread('test.jpg')
plt.imshow(img)

# 保存图片，需放在show()函数之前
plt.savefig('test_gray_1.jpg')
plt.show()
```
